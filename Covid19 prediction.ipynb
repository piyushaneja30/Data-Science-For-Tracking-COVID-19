```python
x= x.reshape(-1,1)
x
```

```python
from sklearn.linear_model import LinearRegression 
regressor= LinearRegression() 
regressor.fit(x,y)
```

```python
yp= regressor.predict(x)
yp
```

```python
plt.scatter(x,y)
plt.plot(x,yp)
plt.show()
```

```python
regressor.score(x,y)*100
```
